```python
filename = "C:/Users/vivek/Desktop/rnn.txt"
rt = open(filename).read()
word = rt
rt
```

```python
A = [ord(c) for c in rt]
```

```python
A
```

```python
# one hot encode sequence
def one_hot_encode(sequence, n_unique=363):
	encoding = list()
	for value in sequence:
		vector = [0 for _ in range(n_unique)]
		vector[value] = 1
		encoding.append(vector)
	return np.array(encoding)
```

```python
enco = one_hot_encode(A, n_unique=363)
```

```python
from numpy import argmax
```

```python
#decode a one hot encoded string
def one_hot_decode(encoded_seq):
	return [argmax(vector) for vector in encoded_seq]
```

```python
deco = one_hot_decode(enco)
```

```python
# create lag inputs
df = DataFrame(enco)
df = concat([df.shift(4), df.shift(3), df.shift(2), df.shift(1), df], axis=1)
df
```

```python
# remove non-viable rows
values = df.values
values = values[5:,:]
values
```

```python
# convert to 3d for input
X = values.reshape(len(values), 5, 363)
X
```

```python
X.shape
```

```python
enco.shape
```

```python
y = enco[4:-1,:]
```

```python
y.shape
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```python
# define model
model = Sequential()
model.add(LSTM(50, batch_input_shape=(39, 5, 363)))
model.add(Dense(363, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
model.fit(X, y, epochs=1000, batch_size=39, verbose=2, shuffle=False)
model.reset_states()
```

```python
yhat = model.predict(X, batch_size=39)
print('Expected:  %s' % one_hot_decode(y))
print('Predicted: %s' % one_hot_decode(yhat))
```
